### CARGAR DATASET LIMPIO

```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df = pd.read_csv('titanic_limpio.csv')
```

###  HISTOGRAMA (Edad)

```python
# crear figura
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# histograma
sns.histplot(df['age'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribución de la Edad')
# barras
sns.countplot(x='sex', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Cantidad por Sexo')

# scatter
sns.scatterplot(x='age', y='fare', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Edad vs Tarifa')

# heatmap
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", ax=axes[1, 1])
axes[1, 1].set_title('Correlaciones')

# mostrar todo
plt.tight_layout()
plt.show()
```
