# Semana3: Actividad 3.3 - Limpieza de datos

### LIMPIEZA Y PREPROCESAMIENTO DE DATOS

```python
import pandas as pd
df = pd.read_csv('train.csv')
print("Dataset cargado correctamente...")
```

### ESTADO ANTES DE LIMPIEZA

```python
print("=== ANTES DE LIMPIEZA ===")
print("\nValores nulos:")
print(df.isnull().sum())

print("\nDuplicados:")
print(df.duplicated().sum())

print("\nTipos de datos:")
print(df.dtypes)
```

### MANEJO DE VALORES NULOS

```python
# Rellenar edad con la media
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Rellenar Embarked con la moda
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Eliminar columna Cabin
df = df.drop(columns=['Cabin'])
```

### ELIMINAR DUPLICADOS

```python
df = df.drop_duplicates()
```

### CONVERTIR TIPOS DE DATOS

```python
df['Survived'] = df['Survived'].astype('category')
```

### ESTANDARIZAR NOMBRES DE COLUMNAS

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

### CREAR NUEVAS COLUMNAS

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
```

### ESTADO DESPUES DE LIMPIEZA

```python
print("\n=== DESPUÉS DE LIMPIEZA ===")
print("\nValores nulos:")
print(df.isnull().sum())

print("\nDuplicados:")
print(df.duplicated().sum())

print("\nTipos de datos:")
print(df.dtypes)

print("\nPrimeras filas:")
df.head()
```
