# Conexión a MongoDB

```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient("mongodb://localhost:27017/")
db = client["empresa"]
collection = db["empleados"]

# Limpiar colección
collection.delete_many({})
```

# CREATE

```python
empleados = [
    {"nombre": "Juan", "departamento": "Ventas", "salario": 8000},
    {"nombre": "Ana", "departamento": "RH", "salario": 9000},
    {"nombre": "Luis", "departamento": "IT", "salario": 12000},
    {"nombre": "Sofia", "departamento": "Ventas", "salario": 8500},
    {"nombre": "Carlos", "departamento": "IT", "salario": 11000},
    {"nombre": "Maria", "departamento": "RH", "salario": 9500},
    {"nombre": "Pedro", "departamento": "Ventas", "salario": 7800},
    {"nombre": "Laura", "departamento": "IT", "salario": 13000},
    {"nombre": "Diego", "departamento": "RH", "salario": 9200},
    {"nombre": "Elena", "departamento": "Ventas", "salario": 8700}
]

collection.insert_many(empleados)
print("Empleados insertados")
```

# READ

```python
consulta = list(collection.find({"departamento": "Ventas"}))
df_read = pd.DataFrame(consulta)

print("Empleados del departamento Ventas:")
df_read
```

# UPDATE

```python
collection.update_one(
    {"nombre": "Juan"},
    {"$set": {"salario": 10000}}
)

print("Salario actualizado")

# Verificar cambio
df_update = pd.DataFrame(list(collection.find({"nombre": "Juan"})))
df_update
```

# DELETE

```python
collection.delete_one({"nombre": "Pedro"})

print("Empleado eliminado")

# Verificar eliminación
df_delete = pd.DataFrame(list(collection.find()))
df_delete
```
