# Crear Datos

```python
import pandas as pd

data = [
    {"nombre": "Juan Pérez", "edad": 25, "ciudad": "Querétaro"},
    {"nombre": "María López", "edad": 30, "ciudad": "CDMX"},
    {"nombre": "Carlos Sánchez", "edad": 22, "ciudad": "Guadalajara"},
    {"nombre": "Ana Torres", "edad": 28, "ciudad": "Monterrey"},
    {"nombre": "Luis Martínez", "edad": 35, "ciudad": "Puebla"}
]

df = pd.DataFrame(data)
```

# Conectar a Mongo y guardar

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["prueba_mongodb"]
collection = db["usuarios"]

# Limpiar por si ya había datos
collection.delete_many({})

# Insertar datos
collection.insert_many(data)
```

# Traer datos de Mongo

```python
data_mongo = list(collection.find())
df_mongo = pd.DataFrame(data_mongo)
df_mongo
```
